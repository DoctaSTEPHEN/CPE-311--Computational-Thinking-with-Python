Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). Hint: Be sure to convert the year column to a number as we did in the previous exercise.

```python
meteorite = pd.read_csv('Meteorite_Landings.csv')
meteorite = meteorite.fillna(0)
meteorite = meteorite.rename(columns={'mass (g)': 'mass'}) #rename column mass(g) to mass for easier access/use
meteorites = lambda x: x.str[6:10]#convert the "year" column to string and get the year through indexing
meteorite.year = meteorites(meteorite.year) #Use lamda, meteorites, to assign "year" column to only years
meteorite.year = (pd.to_numeric(meteorite['year']))#change data type of year column to numeric/float
fall = meteorite[(meteorite['year']>=2005) & (meteorite['year']<=2009)]#filter datas from year 2005-2009

df = {'fall':['found', 'fell'], 
                 'count':[fall.fall[(fall['fall']=='Fell')].count(), fall.fall[(fall['fall']=='Found')].count()], 
                 '95%':[fall.mass[fall['fall']=='Fell'].quantile(.95), fall.mass[fall['fall']=='Found'].quantile(.95)]}
found_vs_fell = pd.DataFrame(df)
found_vs_fell
```

```python
found_vs_fell_pivoted = found_vs_fell.pivot(index = 'fall', columns = 'count', values = '95%')
found_vs_fell_pivoted
```
