Remove characters from the lyrics

```python
import re

newline = lambda x: re.sub(r'\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub(r'\w*\d\w*', ' ', x) # remove alphanumeric words

corpus = beatles.text.map(newline).map(alphanumeric)
corpus.head()
```

Create a document term matrix using Count Vectorizer with the stop words turned on to English

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(beatles.song)
dt.head()
```

Set display options so that the lyrics aren't cut off

```python
pd.set_option('display.max_colwidth', None)
```

```python
beatles[beatles.song=='Imagine'].text
```

Imagine lyrics in Count Vectorizer form


```python
imagine = list(dt.loc['Imagine'])
imagine[:20]
```

Define the cosine similarity calculation

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

Calculate all similarities and sort by the most similar

```python
similarity = [cosine(imagine, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

Create the TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

Calculate all similarities and sort by the most similar using the TF-IDF data

```python
similarity_tfidf = [cosine(imagine, song) for song in X_tfidf]
sorted(list(zip(similarity_tfidf, beatles.song)), reverse=True)[1:6]
```

Calculate the cosine similarity between all combinations of documents

```python

from itertools import combinations

pairs = list(combinations(beatles.song.index, 2)) # all song index combos
pairs_0 = list(combinations(range(len(beatles)), 2)) # all index combos starting with (0,1)
song_pairs = [(beatles.song[a_index], beatles.song[b_index]) for (a_index, b_index) in pairs]
song_pairs[:5]
```

 Calculate the cosine similarity of the Count Vectorizer document-term matrix

```python

results = [cosine(X[a_index], X[b_index]) for (a_index, b_index) in pairs_0]
sorted(zip(results, song_pairs), reverse=True)[:5]
```

Calculate the cosine similarity of the TF-IDF Vectorizer document-term matrix

```python
results = [cosine(X_tfidf[a_index], X_tfidf[b_index]) for (a_index, b_index) in pairs_0]
sorted(zip(results, song_pairs), reverse=True)[:5]
```

Compare the results. Which Vectorizer seems to do a better job?

The results show that songs with similar themes or repeated words, especially about love, tend to have higher similarity scores. The highest similarity occurs when songs have very similar titles or lyrics, while lower scores still indicate shared themes but with more differences in wording.

TF-IDF works better because it does not just count words but also considers how important the words are. It reduces the effect of common words like “love” and focuses more on unique words, so the similarity results are more accurate compared to CountVectorizer.

```python
data = pd.read_csv('/content/drive/MyDrive/CPE 313/McDonald_s_Reviews.csv', encoding='latin1')
data.head()
```

```python
five_str = data[data.rating=='5 stars']
five_str.head()
```

```python
newline = lambda x: re.sub(r'\n', ' ', x)
unknown = lambda x: re.sub(r'[^\x00-\x7F]+', ' ', x)
alphanumeric = lambda x: re.sub(r'\w*\d\w*', ' ', x)

corpus = five_str.review.map(newline).map(unknown).map(alphanumeric)
corpus.head()
```

```python
cv = CountVectorizer(stop_words='english')
X_count = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X_count, columns=cv.get_feature_names_out()).set_index(five_str.reviewer_id)
dt.head()
```

```python
reviewer_64 = list(dt.loc[22155])
reviewer_64[:20]
```

```python
similarity = [cosine(reviewer_64, review_vector) for review_vector in X_count]
sorted(list(zip(similarity, dt.index)), reverse=True)[1:6]
```

```python
cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(five_str.reviewer_id)
dt_tfidf.head()
```

```python
similarity_tfidf = [cosine(reviewer_64, review_vector) for review_vector in X_tfidf]
sorted(list(zip(similarity_tfidf, dt.index)), reverse=True)[1:6]
```

#### Conclusion



To summarize the result of the output, Count Vector was based on frquently used words to find similarity, because of common words like "McDonald's." TF-IDF, on the other hand, was based on important words like "fast" and "food." Although both algorithms detected similar short sentences, TF-IDF is more relevant.

In conclusion, from this activity, I have learned that the similarity of the texts depends on the words used and their importance. The Count Vectorizer method identifies similarity based on the common words between the texts, which may not be the case for the TF-IDF method. The TF-IDF method gives a more accurate result compared to the other method, as it reduces the influence of common words and highlights meaningful terms.
