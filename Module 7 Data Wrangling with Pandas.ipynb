```python
# 2. Add a column to each dataframe, called ticker, 
#    indicating the ticker symbol it is for (Apple's is AAPL, for example).
#    This is how you look up a stock.
#    Each file's name is also the ticker symbol, so be sure to capitalize it.

appl['ticker'] = 'AAPL'
amzn['ticker'] = 'AMZN'
fb['ticker'] = 'FB'
goog['ticker'] = 'GOOG'
nflx['ticker'] = 'NFLX'
```

```python
# 3. Append them together into a single dataframe.
faang = pd.concat([appl,amzn,fb,goog,nflx])
faang
```

```python
faang.to_csv('faang.csv', index = False)
new_df = pd.read_csv('faang.csv')
new_df
```

<h4>Exercise 2</h4>

```python
# With faang, use type conversion to change the date column into a datetime and the volume column into integers.
new_df['date'] = pd.to_datetime(new_df['date'])
new_df['volume'] = pd.to_numeric(new_df['volume'])
new_df.dtypes
```

```python
# Then, sort by date and ticker.
new_df.sort_values('date' and 'ticker', inplace = True)
new_df
```

```python
# Find the seven rows with the highest value for volume.
new_df.sort_values('volume').tail(7)
```

```python
# Right now, the data is somewhere between long and wide format. 
# Use melt() to make it completely long format.
# Hint: date and ticker are our ID variables(they uniquely identify each row).
# We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume
faang_melted = new_df.melt(
    id_vars=['date', 'ticker']
)
faang_melted
```

<h4>Exercise 3</h4>

```python
# Using web scrapping, search for the list of hospitals, their address and contact information.
# Save the list in a new csv file, hospital.csv.

url = 'https://raw.githubusercontent.com/donnemartin/hospital-quality/refs/heads/master/hospital-data.csv'
hospital = pd.read_csv(url, sep = ',', encoding = 'utf-8')
hospital = hospital[['Hospital Name', 'Address 1', 'Phone Number']]
hospital.to_csv('hospital.csv', index = False)
hospital
```

```python
# Using the generated hospital.csv, convert the csv file into pandas dataframe.
# Prepare the data using necessary preprocessing techniques.
Hospital = pd.read_csv('hospital.csv')
Hospital.fillna(0, inplace = True)
Hospital.rename(columns = {'Address 1':'Address', 'Phone Number':'Contact'}, inplace = True)
Hospital
```
